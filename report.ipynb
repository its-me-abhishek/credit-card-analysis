```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
# Load the datasets
credit_card_df = pd.read_csv('./data/credit_card.csv')
customer_df = pd.read_csv('./data/customer.csv')
```

```python
# Display the first few rows of the datasets
print("Credit Card Data (First few rows):")
display(credit_card_df.head())

print("\nCustomer Data (First few rows):")
display(customer_df.head())
```

```python
# Check for missing values
print("\nMissing values in Credit Card Data:")
display(credit_card_df.isnull().sum())

print("\nMissing values in Customer Data:")
display(customer_df.isnull().sum())
```

```python
# Basic statistics for numerical columns
print("\nBasic statistics for Credit Card Data:")
display(credit_card_df.describe())

print("\nBasic statistics for Customer Data:")
display(customer_df.describe())
```

# CREDIT CARD ANALYSIS

```python
## Distribution of Credit Card Categories
plt.figure(figsize=(10, 6))
sns.countplot(data=credit_card_df, x="Card_Category", palette="viridis", hue="Card_Category")
plt.title("Distribution of Credit Card Categories")
plt.xlabel("Card Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
## Revenue by Customer Type
plt.figure(figsize=(10, 6))
revenue_by_customer_type = credit_card_df.groupby('Exp Type')['Total_Trans_Amt'].sum()
sns.barplot(x=revenue_by_customer_type.index, y=revenue_by_customer_type.values)
plt.title('Revenue by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Total Revenue ($)')
plt.show()
```

```python
## Average Utilization Ratio Distribution
plt.figure(figsize=(10, 6))
sns.histplot(credit_card_df['Avg_Utilization_Ratio'], bins=20, kde=True, color='blue')
plt.title('Distribution of Average Utilization Ratio')
plt.xlabel('Average Utilization Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python

## Total Transactions and Interest Earned by Week
plt.figure(figsize=(12, 8))
weekly_trends = credit_card_df.groupby('Week_Num').agg({'Total_Trans_Amt': 'sum', 'Interest_Earned': 'sum'})
sns.lineplot(data=weekly_trends, x=weekly_trends.index, y='Total_Trans_Amt', label='Total Transactions ($)', marker='o')
sns.lineplot(data=weekly_trends, x=weekly_trends.index, y='Interest_Earned', label='Interest Earned ($)', marker='o', color='red')
plt.title('Weekly Trends: Total Transactions and Interest Earned')
plt.xlabel('Week Number')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```

```python
## Delinquent Accounts by Card Category
plt.figure(figsize=(10, 6))
delinquent_by_category = (
    credit_card_df.groupby("Card_Category")["Delinquent_Acc"].sum().reset_index()
)
sns.barplot(
    x="Card_Category",
    y="Delinquent_Acc",
    data=delinquent_by_category,
    palette="plasma",
    hue="Card_Category",
)
plt.title("Delinquent Accounts by Card Category")
plt.xlabel("Card Category")
plt.ylabel("Number of Delinquent Accounts")
plt.xticks(rotation=45)
plt.show()
```

```python
## Boxplot of Credit Limits by Card Category
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=credit_card_df,
    x="Card_Category",
    y="Credit_Limit",
    palette="cubehelix",
    hue="Card_Category",
)
plt.title("Credit Limit Distribution by Card Category")
plt.xlabel("Card Category")
plt.ylabel("Credit Limit ($)")
plt.xticks(rotation=45)
plt.show()
```

# CUSTOMERS ANALYSIS

```python
## Distribution of Customer Age
plt.figure(figsize=(10, 6))
sns.histplot(customer_df['Customer_Age'], bins=20, kde=True, color='blue')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
## Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=customer_df, x="Gender", palette="pastel", hue="Gender")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
## Education Level Distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=customer_df, x='Education_Level', palette='Set2', hue='Education_Level')
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
## Marital Status Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_df, x='Marital_Status', palette='Set3', hue="Marital_Status")
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
## Income Distribution
plt.figure(figsize=(10, 6))
sns.histplot(customer_df['Income'], bins=20, kde=True, color='green')
plt.title('Income Distribution')
plt.xlabel('Income ($)')
plt.ylabel('Frequency')
plt.show()
```

```python

## Customer Satisfaction Score by Age
plt.figure(figsize=(10, 6))
sns.boxplot(data=customer_df, x='Customer_Age', y='Cust_Satisfaction_Score', palette='coolwarm', hue="Customer_Age")
plt.title('Customer Satisfaction Score by Age')
plt.xlabel('Customer Age')
plt.ylabel('Satisfaction Score')
plt.show()
```

```python
## Car Ownership by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=customer_df, x='Car_Owner', hue='Gender', palette='viridis')
plt.title('Car Ownership by Gender')
plt.xlabel('Car Owner')
plt.ylabel('Count')
plt.show()
```
